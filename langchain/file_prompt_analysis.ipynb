````python
# =============================================================================
#   ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
# =============================================================================

# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜
# !pip install PyMuPDF python-docx openai
# pip install python-dotenv

import os
import fitz # PDFìš©
import docx # DOCXìš©
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
````

````python
# =============================================================================
# ğŸ”‘ OpenAI API í‚¤ í™˜ê²½ë³€ìˆ˜ ì„¤ì •
# =============================================================================

# .env íŒŒì¼ ë¡œë“œ (OPENAI_API_KEYë¥¼ í™˜ê²½ë³€ìˆ˜ë¡œ ë“±ë¡)
load_dotenv()

client = OpenAI()
````

````python
# =============================================================================
#   ì²¨ë¶€ íŒŒì¼ ê²½ë¡œ ì„¤ì • ë° í˜•ì‹ í™•ì¸
# =============================================================================

# ì²¨ë¶€í•œ íŒŒì¼ ê²½ë¡œ
file_path = "resume.txt"

# í™•ì¥ì ì¶”ì¶œ í›„ ì†Œë¬¸ìë¡œ ë³€í™˜ (pdf, docx, txt êµ¬ë¶„ìš©)
ext = os.path.splitext(file_path)[-1].lower()
````

````python
# =============================================================================
#   ì²¨ë¶€ íŒŒì¼ì—ì„œ í…ìŠ¤íŠ¸ ì¶”ì¶œ
# =============================================================================

extracted_text = ""

if ext == ".pdf":
    with fitz.open(file_path) as doc:
        for page in doc:
            extracted_text += page.get_text()
elif ext == ".docx":
    doc = docx.Document(file_path)
    for para in doc.paragraphs:
        extracted_text += para.text + "\n"
elif ext == ".txt":
    with open(file_path, "r", encoding="utf-8") as f:
        extracted_text = f.read()
else:
    raise ValueError(f"ì§€ì›í•˜ì§€ ì•ŠëŠ” íŒŒì¼ í˜•ì‹ì…ë‹ˆë‹¤: {ext}")
````

````python
# =============================================================================
#   ì¶”ì¶œëœ í…ìŠ¤íŠ¸ ì¼ë¶€ ë¯¸ë¦¬ë³´ê¸°
# =============================================================================

display(Markdown("ì¶”ì¶œí•œ í…ìŠ¤íŠ¸ ì¼ë¶€ ë¯¸ë¦¬ë³´ê¸°"))
display(Markdown(f"```\n{extracted_text[:1000]}...\n```"))
````

````python
# =============================================================================
#   Ollamaë¡œ ë³´ë‚¼ í”„ë¡¬í”„íŠ¸ ì‘ì„±
# =============================================================================

# Ollama í”„ë¡¬í”„íŠ¸
prompt = f"""
ë„ˆëŠ” ì „ë¬¸ ì¸ì‚¬ ë‹´ë‹¹ìì•¼.

ì•„ë˜ëŠ” ì§€ì›ìì˜ ì´ë ¥ì„œ ë‚´ìš©ì´ì•¼. ë‚´ìš©ì„ ë¶„ì„í•˜ê³  ì•„ë˜ ì§ˆë¬¸ì— ë‹µí•´ì¤˜:
1) ì§€ì›ìì˜ í•µì‹¬ ê²½ë ¥ê³¼ ê¸°ìˆ  ì—­ëŸ‰ì€?
2) ì§€ì›ìì˜ ê²½ë ¥ ì¤‘ ê°€ì¥ ê°•ì ì´ ë˜ëŠ” ë¶€ë¶„ì€?
3) ê°œì„ í•˜ê±°ë‚˜ ë³´ì™„í•˜ë©´ ì¢‹ì„ ì ì€?

[ì´ë ¥ì„œ ë‚´ìš©]
-----------------
{extracted_text}
-----------------

â€» ë‹µë³€ì€ Markdown í˜•ì‹ìœ¼ë¡œ ì‘ì„±í•´ì¤˜.
"""
````

````python

# =============================================================================
#   Ollama HTTP API í˜¸ì¶œ
# =============================================================================

response = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": "ë‹¹ì‹ ì€ ì „ë¬¸ ì¸ì‚¬ ë‹´ë‹¹ìì…ë‹ˆë‹¤."},
        {"role": "user", "content": prompt}
    ]
)

analysis = response.choices[0].message.content
````

````python

# =============================================================================
#   ë¶„ì„ ê²°ê³¼ ì¶œë ¥
# =============================================================================

display(Markdown("GPT ë¶„ì„ ê²°ê³¼ (OpenAI)"))
display(Markdown(analysis))
````
